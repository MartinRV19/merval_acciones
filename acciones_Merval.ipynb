# Obtener datos


```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import seaborn as sns
import numpy as np
import yfinance as yf
import pandas as pd
```

```python
lista_merval = ['ALUA.BA'
,'BBAR.BA'
,'BMA.BA'
,'BYMA.BA'
,'CEPU.BA'
,'COME.BA'
,'CRES.BA'
,'CVH.BA'
,'EDN.BA'
,'GGAL.BA'
,'HARG.BA'
,'LOMA.BA'
,'MIRG.BA'
,'PAMP.BA'
,'SUPV.BA'
,'TECO2.BA'
,'TGNO4.BA'
,'TGSU2.BA'
,'TRAN.BA'
,'TXAR.BA'
,'VALO.BA'
,'YPFD.BA']

datos = yf.download(lista_merval)['Close']
```

```python
df = datos.copy()
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df.head()
```

Filtrar por fechas que interesan (2018 a 2022)

```python
df = df[(df.index>='2018-01-01')&(df.index<'2023-01-01')]

columnas = df.columns

lista_nueva_columnas=[]
for i in columnas:
  ticker = i.replace('.BA','')
  lista_nueva_columnas.append(ticker)

df.columns = lista_nueva_columnas
df
```

```python
# Por errores de actualización de datos diarios no traer [BBAR,HARG,TXAR]
df.drop(['BBAR','HARG','TXAR'],axis=1,inplace=True)
```

# Cotización

```python
fig, ax = plt.subplots(figsize=(14,6))

for i in df.columns:
  ticker = i 
  valor = round(df.tail(1)[f'{i}'][0])
  ticker_valor = f'{ticker} ${valor}'
  ax.plot(df[f'{i}'],label=ticker_valor)

ax.set_title('Cotización acciones Merval y valor a 30-12-2022',size=16)

pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.9, pos.height])
ax.legend(loc='center right', bbox_to_anchor=(1.2, 0.5))

plt.show()
```

# Retorno diario

```python
df_pct_change = df.copy()

columnas = df_pct_change.columns

for i in columnas:
  df_pct_change[i] = df_pct_change[i].pct_change()

df_pct_change.dropna(inplace=True)
df_pct_change.head()
```

```python
corr = df_pct_change.corr(method='spearman')

mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(16, 16))

sns.heatmap(corr, annot = True, mask=mask, cmap="YlGnBu", center=0,
            square=True, linewidths=.5, fmt='.2f')

ax.set_title('Correlación entre acciones',size=20)

plt.show()
```

## Top Subas de Retorno Diario

```python
top_subas_diaria = pd.DataFrame({'Fecha':df_pct_change.idxmax(),'Ticker':df_pct_change.max().index,'Porcentaje':df_pct_change.max()}).set_index('Fecha').sort_values('Porcentaje',ascending=False)[:10]
top_subas_diaria['Porcentaje'] = top_subas_diaria['Porcentaje'].mul(100)
top_subas_diaria
```

Se elimina BYMA por error de actualización de datos diarios

```python
# BYMA outlier por falta de actualización de datos en 2018
top_subas_diaria = top_subas_diaria.loc[top_subas_diaria['Ticker'] !='BYMA']
top_subas_diaria
```

## Top Caídas de Retorno Diario

```python
top_caida_diaria = pd.DataFrame({'Fecha':df_pct_change.idxmin(),'Ticker':df_pct_change.min().index,'Porcentaje':df_pct_change.min()}).set_index('Fecha').sort_values('Porcentaje')[:10]
top_caida_diaria['Porcentaje'] = top_caida_diaria['Porcentaje'].mul(100)
top_caida_diaria = top_caida_diaria.loc[top_caida_diaria['Ticker'] !='BYMA']
top_caida_diaria
```

# Retorno anual

```python
df_pct_change_year = df.copy()

columnas = df_pct_change_year.columns
años = df_pct_change_year.index.year.unique()

lista = []

for i in años:
  anio = i
  data = df_pct_change_year[df_pct_change_year.index.year==anio]
  for i in columnas:
    ticker = i
    valor_inicial = data[f'{i}'][0]
    valor_final = data[f'{i}'][-1]
    retorno_anual = round((valor_final - valor_inicial) / valor_inicial,4)
    value_dict = {'Fecha':anio,'Ticker':ticker,'Porcentaje':retorno_anual}
    lista.append(value_dict)

df_yoy = pd.DataFrame.from_dict(lista)
df_yoy['Porcentaje'] = df_yoy['Porcentaje'].mul(100) 
df_yoy
```

Errores de actualización de datos diarios en 2018 BYMA y en 2022 TGSU2

2018	BYMA	1012.94

2022	TGSU2	337.51
